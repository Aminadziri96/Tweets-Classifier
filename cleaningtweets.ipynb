### <a style="color:#D1A8B9" font=" Bookman"  size="75" >Definir la fonction qui elimine le ponctuation</a>

```python
import re
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

df['text'] = df['text'].apply(lambda x: remove_punct(x))
df.head(10)
```

## <a style="color:#C37D58" font=" Bookman"  size="75" > 2-Tokenization </a>


```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

df['text'] = df['text'].apply(lambda x: tokenization(x.lower()))
df.head()
```

```python
stopword = nltk.corpus.stopwords.words('english')
stopword.extend(['b','yr', 'year', 'woman', 'man', 'girl','boy','one', 'two', 'sixteen', 'yearold', 'fu', 'weeks', 'week', 'treatment', 'associated', 'patients', 'may','day', 'case','old'])
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
df['text'] = df['text'].apply(lambda x: remove_stopwords(x))
df.head(10)
```

## <a style="color:#C37D58" font=" Bookman"  size="75" > 4-Stemming and Lammitization </a>



```python
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

df['text'] = df['text'].apply(lambda x: stemming(x))
df.head()
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

df['text'] = df['text'].apply(lambda x: lemmatizer(x))
df.head()
```

```python
def clean_text(text):
    text_lc = "".join([word.lower() for word in text if word not in string.punctuation]) # remove puntuation
    text_rc = re.sub('[0-9]+', '', text_lc)
    tokens = re.split('\W+', text_rc)    # tokenization
    text = [ps.stem(word) for word in tokens if word not in stopword]  # remove stopwords and stemming
    return text
df['text'] = df['text'].apply(lambda x: clean_text(x))
df.head()
```

#  <a style="color:#C37D58" font=" Bookman"  size="75" > Autre methode data cleaning and preprocessing using regular expressions</a>



```python
import re

def nlp_pipeline(text):

    text = text.lower()
    text = text.replace('\n', ' ').replace('\r', '')
    text = ' '.join(text.split())
    text = re.sub(r"[A-Za-z\.]*[0-9]+[A-Za-z%°\.]*", "", text)
    text = re.sub(r"(\s\-\s|-$)", "", text)
    text = re.sub(r"[,\!\?\%\(\)\/\"]", "", text)
    text = re.sub(r"\&\S*\s", "", text)
    text = re.sub(r"\&", "", text)
    text = re.sub(r"\+", "", text)
    text = re.sub(r"\#", "", text)
    text = re.sub(r"\$", "", text)
    text = re.sub(r"\£", "", text)
    text = re.sub(r"\%", "", text)
    text = re.sub(r"\:", "", text)
    text = re.sub(r"\@", "", text)
    text = re.sub(r"\-", "", text)
    text = re.sub(r"\...", "", text)
    text = re.sub(r"\'", "", text)

    return text
```

```python
import pandas as pd
tweet  = pd.read_csv('csv/economy_tweets.csv')


corpus = tweet['text']
corpus_clean = corpus.apply(nlp_pipeline)
```
